```python
import ast
```

```python
def convert(obj):
    L=[]
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
movies['genres'] = movies['genres'].apply(convert)
movies['keywords'] = movies['keywords'].apply(convert)
movies.head()
```

```python
def convert3(obj):
    L = []
    counter = 0
    for i in ast.literal_eval(obj):
        if counter != 3:
            L.append(i['name'])
            counter+=1
        else:
            break
        return L
```

```python
movies['cast'] = movies['cast'].apply(convert3)
```

```python
movies.head()
```

```python
def fetch_director(obj):
    L = []
    for i in ast.literal_eval(obj):
        if i['job']=='Director':
            L.append(i['name'])
    return L
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
movies.head()
```

```python
movies['overview'][0]
```

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
movies.head()
```

```python
movies['genres'] = movies['genres'].apply(lambda x: [i.replace(" ","")for i in x] if x is not None else [])
movies['keywords'] = movies['keywords'].apply(lambda x: [i.replace(" ","")for i in x] if x is not None else [])
movies['cast'] = movies['cast'].apply(lambda x: [i.replace(" ","")for i in x] if x is not None else [])
movies['crew'] = movies['crew'].apply(lambda x: [i.replace(" ","")for i in x] if x is not None else [])
```

```python
movies.head()
```

```python
movies['tags'] = movies['overview']+movies['genres']+movies['keywords']+movies['cast']+movies['crew']
```

```python
movies.head()
```

```python
new = movies[['movie_id','title','tags']]
```

```python
new.head()
```

```python
new['tags'] = new['tags'].apply(lambda x:' '.join(x))
```

```python
new.head()
```

```python
new['tags'] = new['tags'].apply(lambda x:x.lower())
```

```python
new.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features = 5000, stop_words='english')
```

```python
cv.fit_transform(new['tags']).toarray().shape
```

```python
vectors = cv.fit_transform(new['tags']).toarray()
```

```python
vectors[0]
```

```python
len(cv.get_feature_names())
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
new['tags'] = new['tags'].apply(stem)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
cosine_similarity(vectors)
```

```python
cosine_similarity(vectors).shape
```

```python
similarity = cosine_similarity(vectors)
```

```python
similarity[0]
```

```python
similarity[0].shape
```

```python
sorted(list(enumerate(similarity[0])), reverse = True, key = lambda x:x[1])[1:6]
```

```python
def recommend(movie):
    movie_index = new[new['title']==movie].index[0]
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)), reverse= True, key= lambda x:x[1])[1:6]
    
    for i in movies_list:
        print(new.iloc[i[0]].title)
```

```python
recommend('Avatar')
```

```python
recommend('The Avengers')
```

```python
recommend('Ant-Man')
```
